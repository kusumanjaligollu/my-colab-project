```python

!pip install ipywidgets --quiet

import ipywidgets as widgets
from IPython.display import display, clear_output

# ----------- LOGO -----------
def print_logo():
    print("="*60)
    print("🎓✨ [ Welcome to ANITS  Engineering College Chatbot ] ✨🎓")
    print(" ANITS official website: https://www.anits.edu.in")
    print("="*60)

# ----------- BOT -----------
def chatbot_buttons():
    print_logo()
    print("\n🤖 Hello! I am your college guide bot. Please select an option:")

    # Define Buttons
    button_courses = widgets.Button(description="📚 Courses", button_style='success')
    button_admissionprocess = widgets.Button(description="📝 Admissionprocess ", button_style='info')
    button_fees = widgets.Button(description="💰 Fees", button_style='warning')
    button_hostel = widgets.Button(description="🏠 Hostel", button_style='primary')
    button_location = widgets.Button(description="📍 Location", button_style='danger')
    button_placements = widgets.Button(description="💼 Placements", button_style='success')
    button_exit = widgets.Button(description="❌ Exit", button_style='danger')

    # Group Buttons
    buttons = [button_courses, button_admissionprocess, button_fees,
               button_hostel, button_location, button_placements, button_exit]

    output = widgets.Output()
    display(widgets.VBox(buttons), output)

    def on_button_clicked(b):
        with output:
            clear_output()
            if b.description.startswith("📚"):
                print("\n📚 We offer B.Tech-[CSE,ECE,EEE,CHEM,MECH,CIVIL] \n-M.Tech in various specilizations ,\n ~ MCA and MBA programs")
            elif b.description.startswith("📝"):
                print("\n📝 Admissions are open from May to July.here are two ways to get a admission in ANITS\n1.Through APEMACET RANK[EMACET CODE: ANIL]. \n2.Management quota( Visit our website for online admissions).")
            elif b.description.startswith("💰"):
                print("\n💰 The average fees for B.Tech are around 1.5 lakh per year and also it  varies how student get the seat \n ~ through EMACET free seat FEES would probably be 21,500 i.e building fund ")
            elif b.description.startswith("🏠"):
                print("\n🏠 \n~ Seperate Hostels for boys and girls \n~ AC & NON AC Rooms are available \n~ hygenic food \n ~24/7 security.")
            elif b.description.startswith("📍"):
                print("\n📍 Sangivalasa ,bheemunipatnam mandal,visakhapatnam \n ~ It is 27 km from vizag city \n ~ it is 25 km from viziangaram .")
            elif b.description.startswith("💼"):
                print("\n💼 Our placement Record \n Top Recrutiers : Infosys,TCS,Wipro,Tech Mahindra  cell has tie-ups with top MNCs💻,\n~ Highest Packages 44 LPA \n with an average package of 6 LPA.")
            elif b.description.startswith("❌"):
                print("\n👋 Thank you for chatting with me! I hope this chat gives you information about our college, Have a great day😊!")
                for btn in buttons:
                    btn.close()


    # Add Click Function to Buttons
    for button in buttons:
        button.on_click(on_button_clicked)

# ----------- MAIN -----------
chatbot_buttons()
```
